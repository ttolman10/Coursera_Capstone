```python
#Using Pandas libraries to parse Wikipedia Website for Toronto data
import pandas as pd
from pandas.io.html import read_html
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
#After inspecting site html, grabbing the table with class = "wikitable sortable" and creating df
webtable = read_html(page, attrs={"class":"wikitable sortable"})
df = webtable[0]
df.head(15)
```

```python
#removing rows where Borough = "Not assigned"
df = df[df.Borough != "Not assigned"]
df.head(10)
```

```python
#finding rows with 'Neighbourhood' value set to 'Not assigned'
df[df['Neighbourhood'] == 'Not assigned']['Postcode']
```

```python
#Setting Neighborhood on index row 7 to 'Queen's Park'
df.loc[7,'Neighbourhood'] = 'Queen\'s Park'
```

```python
df.head(15)
```

```python
#grouping Postcode by Neighborhood
df_clean = df.groupby("Postcode").agg(lambda x:','.join(set(x)))
df_clean
```

```python
#resetting index values
df_clean.reset_index()
```

```python
#getting shape of dataframe
df_clean.shape
```

Getting Geospacial Data from .csv file

```python
#reading Geospatial_Coordinates.csv and assigning to dataframe
df_geodata = pd.read_csv('Geospatial_Coordinates.csv')
df_geodata.head()
```

```python
#updating df_geodata column header to match df_clean
df_geodata.rename(columns = {'Postal Code':'Postcode'}, inplace = True)
df_geodata.head()
```

```python
type(df_geodata)
```

```python
#merging dataframes along Postal Code
df_merged = pd.merge(df_clean, df_geodata, on='Postcode')
df_merged.head()
```
